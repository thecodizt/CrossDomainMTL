```python
import geopandas as gpd
```

```python
basins_layer = gpd.read_file('./GeoSpatial/geospatial.gpkg', layer='basins')

basins_layer.drop(columns=['name'], inplace=True)

basins_layer
```

```python
sites_layer = gpd.read_file('./GeoSpatial/geospatial.gpkg', layer='sites')

sites_layer.rename(columns={'geometry': 'location'}, inplace=True)

sites_layer
```

```python
cross_sectional_data = pd.merge(metadata, basins_layer, on='site_id')
cross_sectional_data = pd.merge(cross_sectional_data, sites_layer, on='site_id')

cross_sectional_data
```

```python
cross_sectional_data.to_csv('./CSV/cross-sectional-data.csv')
```
